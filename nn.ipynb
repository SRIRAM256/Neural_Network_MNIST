```python
def intialize(x):

    n=x.shape[0]
    w1=np.random.rand(10,n)-0.5
    b1=np.random.rand(10,1)-0.5
    w2=np.random.rand(10,10)-0.5
    b2=np.random.rand(10,1)-0.5
    return w1,b1,w2,b2

def Relu(x):
    return np.maximum(x,0)

def softmax(x):
    total=sum(np.exp(x))
    return np.exp(x)/total

def forward(x,w1,b1,w2,b2):
    a0=x
    z1=w1.dot(a0)+b1
    a1=Relu(z1)
    z2=w2.dot(a1)+b2
    a2=softmax(z2)
    return z1,a1,z2,a2

def num_encode(Y):
  Y=Y.astype(int)
  one_hot_Y = np.zeros((Y.size, Y.max() + 1))
  one_hot_Y[np.arange(Y.size), Y] = 1
  return one_hot_Y.T


def backward(w1,b1,w2,b2,z1,a1,z2,a2,x,y):
    m=x.shape[1]
    y=num_encode(y)

    dz2=a2-y
    dw2=1/m*(dz2.dot(a1.T))
    db2=1/m*(np.sum(dz2))

    dz1=w2.T.dot(dz2)*(z1>0)
    dw1=1/m*(dz1.dot(x.T))
    db1=1/m*(np.sum(dz1))

    return dw1,db1,dw2,db2



def updata_params(w1,b1,w2,b2,dw1,db1,dw2,db2,alpha):
    w1=w1-alpha*dw1
    b1=b1-alpha*db1
    w2=w2-alpha*dw2
    b2=b2-alpha*db2

    return w1,b1,w2,b2


def predict(a2):
    return np.argmax(a2,0)


def accuracy(y_cap,y):
    return np.sum(y_cap==y)/y.size


def train(x,y,alpha,epochs):
    w1,b1,w2,b2=intialize(x)

    for i in range(epochs):
        z1,a1,z2,a2=forward(x,w1,b1,w2,b2)
        dw1,db1,dw2,db2=backward(w1,b1,w2,b2,z1,a1,z2,a2,x,y)
        w1,b1,w2,b2=updata_params(w1,b1,w2,b2,dw1,db1,dw2,db2,alpha)

        if (i%20)==0:
            print("Iteration number: ", i)
            print("Accuracy = ", accuracy(predict(a2), y))

    return w1,b1,w2,b2
```

```python
w1,b1,w2,b2=train(x_train,y_train,0.1,500)
```
